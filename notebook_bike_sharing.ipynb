# Cleaning Data

1. Menghapus tabel yang tidak diperlukan.

Sebelum melanjutkan ke tahapan-tahapan berikutnya, kita akan menghapus tabel hr. Alasannya adalah karena tabel tersebut tidak relevan dengan pertanyaan bisnis yang telah ditetapkan. Selain itu, tabel hr dan day hanya berbeda pada kolom hr, sementara kolom lainnya sama.

```python
del hr_df
```

2. Menghapus beberapa kolom yang tidak akan digunakan.

Berdasarkan pertanyaan bisnis yang telah ditetapkan, kolom-kolom berikut akan dihapus: <br>
<li> instant: Indeks catatan ini tidak memberikan informasi tambahan yang signifikan.</li>
<li>windspeed: Tidak ada relevansi antara kecepatan angin dan jumlah penyewa sepeda berdasarkan pertanyaan bisnis yang telah ditetapkan.</li>

```python
drop_col = ['instant', 'windspeed']

for i in day_df.columns:
  if i in drop_col:
    day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
```

3. Memodifikasi beberapa detail pada kolom.

Karena nama-nama kolom tersebut kurang sesuai dengan preferensi saya, maka saya akan mengubahnya. Selain itu, saya lebih menyukai tipe data dalam bentuk kategorikal, sehingga beberapa kolom yang sebelumnya berupa angka akan saya ubah. Ada juga beberapa tipe data yang tidak sesuai, dan akan saya perbaiki.

```python
# Mengubah beberapa detail tentang kolom
# Mengubah nama judul kolom
day_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_cond',
    'cnt': 'count'
}, inplace=True)

day_df.head()
```

```python
# Mengubah angka menjadi keterangan
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['weather_cond'] = day_df['weather_cond'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})
```

```python
# Mengubah tipe data ke datetime
day_df['dateday'] = pd.to_datetime(day_df.dateday)

# Mengubaha tipe data ke categorical
day_df['season'] = day_df.season.astype('category')
day_df['year'] = day_df.year.astype('category')
day_df['month'] = day_df.month.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weather_cond'] = day_df.weather_cond.astype('category')

day_df.head()
```

8. Menganalisis hubungan antara penyewa casual, registered, dan total penyewa (count).

# Visualization & Explanatory Analysis

**Pertanyaan 1 : Apakah kondisi cuaca (seperti suhu, dan kelembapan) mempengaruhi jumlah pengguna sepeda dalam satu hari?**

```python
# Membuat interval (bins) untuk suhu dan kelembapan
day_df['temp_bins'] = pd.cut(day_df['temp'], bins=10)
day_df['hum_bins'] = pd.cut(day_df['hum'], bins=10)

plt.figure(figsize=(15, 6))

# Plot untuk suhu (temp) dengan binning
plt.subplot(1, 2, 1)
sns.barplot(
    x='temp_bins',
    y='count',
    data=day_df,
    hue='temp_bins',  # Menetapkan 'hue' agar sesuai dengan 'x'
    dodge=False,      # Menghindari pemisahan berdasarkan hue
    legend=False,     # Menyembunyikan legenda karena hanya satu kategori
    errorbar=None)
plt.title('Distribusi Pengguna Sepeda Berdasarkan Rentang Suhu', fontsize=14)
plt.xlabel('Rentang Suhu (Binned)', fontsize=12)
plt.ylabel('Total Pengguna Sepeda', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Menambahkan grid horizontal untuk keterbacaan

# Plot untuk kelembapan (humidity) dengan binning
plt.subplot(1, 2, 2)
sns.barplot(
    x='hum_bins',
    y='count',
    data=day_df,
    hue='hum_bins',  # Menetapkan 'hue' agar sesuai dengan 'x'
    dodge=False,     # Menghindari pemisahan berdasarkan hue
    legend=False,    # Menyembunyikan legenda karena hanya satu kategori
    errorbar=None)
plt.title('Distribusi Pengguna Sepeda Berdasarkan Rentang Kelembapan', fontsize=14)
plt.xlabel('Rentang Kelembapan (Binned)', fontsize=12)
plt.ylabel('Total Pengguna Sepeda', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Menambahkan grid horizontal untuk keterbacaan

plt.tight_layout()
plt.show()
```

Kesimpulan :
1. Suhu dan kelembapan memang memiliki pengaruh signifikan terhadap jumlah pengguna sepeda. Pengguna sepeda cenderung lebih banyak pada kondisi suhu yang sedang hingga hangat, serta pada tingkat kelembapan yang sedang.
2. Kondisi ekstrem, baik suhu yang terlalu dingin atau terlalu panas, serta kelembapan yang sangat rendah atau sangat tinggi, cenderung mengurangi jumlah pengguna sepeda.

**Bagaimana perbedaan jumlah pengguna sepeda pada hari kerja (working day) dibandingkan dengan hari libur dan akhir pekan (holiday dan weekend)?**

```python
plt.figure(figsize=(12, 6))

# Data untuk perbandingan
data = {
    'Jenis Hari': ['Hari Kerja', 'Libur/Akhir Pekan'],
    'Jumlah Pengguna': [
        day_df[day_df['workingday'] == 1]['count'].mean(),  # Rata-rata jumlah pengguna sepeda pada hari kerja
        day_df[(day_df['workingday'] == 0) | (day_df['holiday'] == 1)]['count'].mean()  # Rata-rata jumlah pengguna sepeda pada hari libur dan akhir pekan
    ]
}

# Konversi data ke dalam DataFrame
df_plot = pd.DataFrame(data)

# Plot dengan menetapkan 'hue' dan menghapus 'palette'
sns.barplot(
    x='Jenis Hari',
    y='Jumlah Pengguna',
    data=df_plot,
    hue='Jenis Hari',  # Mengatur hue agar sesuai dengan x
    dodge=False,       # Menghindari pemisahan berdasarkan hue
    legend=False,      # Menyembunyikan legenda karena hanya satu kategori
    errorbar=None)

plt.title('Perbandingan Jumlah Pengguna Sepeda pada Hari Kerja vs. Hari Libur/Akhir Pekan', fontsize=14)
plt.xlabel('Jenis Hari', fontsize=12)
plt.ylabel('Rata-rata Jumlah Pengguna Sepeda', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Menambahkan grid horizontal untuk memudahkan pembacaan
plt.show()
```

Kesimpulan : Sepeda tetap menjadi pilihan populer baik pada hari kerja maupun saat libur atau akhir pekan. Namun, terdapat sedikit keunggulan dalam penggunaan sepeda pada hari kerja dibandingkan dengan hari libur atau akhir pekan, yang menunjukkan bahwa sepeda lebih sering digunakan sebagai alat transportasi utama untuk bekerja atau kegiatan harian daripada untuk rekreasi pada akhir pekan.
